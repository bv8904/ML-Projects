#Importing all the important and necessary Libraries

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

#Uploading the dataset

```python
dataset=pd.read_csv('/content/mail_data.csv')
```

```python
#prints the no.of rows and columns in dataset
dataset.shape
```

```python
#head() prints the first five rows of dataset
dataset.head()
```

#Checking for null values

```python
dataset.isnull().sum()
#Handling null values if present
dataset.fillna(' ')
```

```python
Y=dataset['Category']
X=dataset['Message']
#data
print(X)
#labels
print(Y)
```

#Training and Testing the data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
print(X.shape)
print(X_train.shape)
print(X_test.shape)
print(Y.shape)
print(Y_train.shape)
print(Y_test.shape)
```

#Tfidf Vectorizer

```python
tfidf=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
X_train_features=tfidf.fit_transform(X_train)
X_test_features=tfidf.transform(X_test)
print(X_train_features)
print(X_test_features)
```

#Logistic Regression

```python
lrmodel=LogisticRegression()
#Trainig the model
lrmodel.fit(X_train_features,Y_train)
#predicting labels for the training data
prediction=lrmodel.predict(X_train_features)
#evaluating the accuracy of the model for training data
train_accuracy=accuracy_score(Y_train,prediction)
#printing training accuracy
print("Accuracy of the training dataset ",train_accuracy)
#predicting labels for the testing data
prediction=lrmodel.predict(X_test_features)
#evaluating the accuracy of the model for testing data
test_accuracy=accuracy_score(Y_test,prediction)
#printing testing accuracy
print("Accuracy of the testing dataset  ",test_accuracy)
```

#Checking the model

```python
input=["I'm gonna be home soon and i don't want to talk about this stuff anymore tonight, k? I've cried enough today."]
input_feature=tfidf.transform(input)
prediction=lrmodel.predict(input_feature)
if prediction[0]=='S':
  print("Spam")
else:
  print("Ham")
```
