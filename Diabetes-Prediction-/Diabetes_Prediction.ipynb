#Importing all the important and necessary Libraries

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
import warnings
```

#Uploading the dataset
1 --> Diabetic
0 --> Non-Diabetic

```python
dataset=pd.read_csv('/content/augmented_diabetes.csv')
#head() prints the first five rows of dataset
dataset.head()
```

```python
#prints the no.of rows and columns in dataset
dataset.shape
```

```python
#describe() gives all the statistical values
dataset.describe()
```

```python
#checking no.of diabetic and non-diabetic data points
dataset['Outcome'].value_counts()
```

#Checking for null values

```python
dataset.isnull().sum()
```

#Data Standardization

```python
Y=dataset['Outcome']
X=dataset.drop(columns='Outcome',axis=1)
print(X)
print(Y)
```

```python
data_standardize=StandardScaler()
X=data_standardize.fit_transform(X)
print(X)
print(Y)
```

#Training and Testing the data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2,stratify=Y)
print(X.shape)
print(X_train.shape)
print(X_test.shape)
print(Y.shape)
print(Y_train.shape)
print(Y_test.shape)
```

#Support Vector Machine

```python
model=svm.SVC(kernel='linear')
#Trainig the model
model.fit(X_train,Y_train)
#predicting labels for the training data
train_prediction=model.predict(X_train)
#evaluating the accuracy of the model for training data
train_accuracy=accuracy_score(Y_train,train_prediction)
print('Accuracy of training data:',train_accuracy)
#predicting labels for the testing data
test_prediction=model.predict(X_test)
#evaluating the accuracy of the model for testing data
test_accuracy=accuracy_score(Y_test,test_prediction)
print('Accuracy of testing data:',test_accuracy)
```

#Checking the model

```python
input=[5.0,116.0,74.0,0.0,0.0,25.6,0.201,30.0]
arr_input=np.asarray(input)
#to standardize the input data
standardize=data_standardize.transform(arr_input.reshape(1,-1))
ans=model.predict(standardize)
warnings.filterwarnings('ignore', message='X does not have valid feature names')
if(ans[0]==1):
  print('Diabetic')
else:
  print('Non-Diabetic')
```
